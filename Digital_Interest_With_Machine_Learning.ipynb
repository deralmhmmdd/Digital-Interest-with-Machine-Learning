```python
import pandas as pd
import numpy as np
import random
from faker import Faker
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Menghasilkan data Dummy ataupun data Palsu yang akan di gunakan dalam proses Preprocessing
fake = Faker()
```

```python
# Generate Dummy Data
def generate_dummy_data(num_entries):
  data = []
  for _ in range(num_entries):
    nama_lokasi = fake.city()
    jam_login = fake.time(pattern ='%H:%M:%S')
    name = fake.name()
    email = fake.email()
    no_telp = fake.phone_number()
    tahun_lahir = random.randint(1960,2015)
    merk_hp = random.choice(['Samsung','Apple', 'Xiaomi','Huawei','Iphone','Vivo','Oppo','Infinix'])
    digital_interest = random.choice(['Sosial Media','E-Commerce','News','Gaming','Entertaiment'])
    location_type = random.choice (['indonesia',"Singapore",'India',"AS",'Jepang', 'Korea Selatan','China','Ruisa'])
    data.append([nama_lokasi, jam_login, name, email, no_telp, tahun_lahir, merk_hp, digital_interest, location_type])

  return data

```

```python
# Menuliskan hasil data Dummy di dalam bentuk File CSV
def write_to_csv(filename, data):
  with open(filename, 'w',newline='') as file:
    writer =csv.writer(file)
    writer.writerow(['nama_lokasi', 'jam_login', 'name, email', 'no_telp', 'tahun_lahir', 'merk_hp', 'digital_interest', 'location_type'])
    writer.writerow(data)
    for entry in data:
            writer.writerow(entry)
```

```python
# Menghasilkan Data Dummy sebanyak 1000 baris
dummy_data = generate_dummy_data(1000)
```

```python
import pandas as pd

# Membuat DataFrame dari data dummy
df = pd.DataFrame(dummy_data, columns=['nama_lokasi', 'jam_login', 'name', 'email', 'no_telp', 'tahun_lahir', 'merk_hp', 'digital_interest', 'location_type'])

# Menyimpan DataFrame ke file CSV
df.to_csv('dummy.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# pivot_table dari tabel
merk_hp = ['Apple', 'Huawei', 'Infinix', 'Iphone', 'Oppo', 'Samsung', 'Vivo', 'Xiaomi']
location_types = ['AS', 'China', 'India', 'Jepang', 'Korea Selatan', 'Rusia', 'Singapore', 'Indonesia']

pivot_table = {
    'AS': [17, 18, 9, 19, 17, 14, 18, 18],
    'China': [14, 16, 27, 10, 11, 13, 13, 17],
    'India': [20, 12, 20, 14, 18, 16, 21, 18],
    'Jepang': [19, 16, 14, 9, 15, 11, 18, 16],
    'Korea Selatan': [11, 7, 17, 11, 17, 27, 14, 16],
    'Rusia': [15, 19, 12, 15, 15, 13, 18, 15],
    'Singapore': [13, 15, 11, 15, 20, 16, 17, 12],
    'Indonesia': [20, 14, 21, 12, 15, 15, 22, 12]
}

# Bar Chart
plt.figure(figsize=(10, 6))
for location_type in location_types:
    plt.bar(merk_hp, pivot_table[location_type], label=location_type)

plt.title('Jumlah Penggunaan Merek HP berdasarkan Location Type')
plt.xlabel('Merk HP')
plt.ylabel('Jumlah Penggunaan')
plt.xticks(rotation=45)
plt.legend(title='Location Type')
plt.tight_layout()
plt.show()

# Heatmap
df = pd.DataFrame(pivot_table, index=merk_hp)
plt.figure(figsize=(10, 6))
sns.heatmap(df, cmap='YlGnBu', annot=True, fmt='g')
plt.title('Heatmap Penggunaan Merek HP berdasarkan Location Type')
plt.xlabel('Location Type')
plt.ylabel('Merk HP')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Stacked Bar Chart
plt.figure(figsize=(10, 6))
bottom = None
for location_type in location_types:
    if bottom is None:
        plt.bar(merk_hp, pivot_table[location_type], label=location_type)
        bottom = pivot_table[location_type]
    else:
        plt.bar(merk_hp, pivot_table[location_type], bottom=bottom, label=location_type)
        bottom = [bottom[i] + pivot_table[location_type][i] for i in range(len(merk_hp))]

plt.title('Stacked Bar Chart Penggunaan Merek HP berdasarkan Location Type')
plt.xlabel('Merk HP')
plt.ylabel('Jumlah Penggunaan')
plt.xticks(rotation=45)
plt.legend(title='Location Type')
plt.tight_layout()
plt.show()

# Line Chart
plt.figure(figsize=(10, 6))
for location_type in location_types:
    plt.plot(merk_hp, pivot_table[location_type], label=location_type)

plt.title('Line Chart Penggunaan Merek HP berdasarkan Location Type')
plt.xlabel('Merk HP')
plt.ylabel('Jumlah Penggunaan')
plt.xticks(rotation=45)
plt.legend(title='Location Type')
plt.tight_layout()
plt.show()

# Pie Chart
plt.figure(figsize=(15, 15))
explode = (0.1, 0, 0, 0, 0, 0, 0, 0)  # explode 1st slice
for i, location_type in enumerate(location_types):
    plt.subplot(3, 3, i+1)
    plt.pie(pivot_table[location_type], labels=merk_hp, explode=explode, autopct='%1.1f%%', shadow=True)
    plt.title(location_type)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.stats import norm

# Data E-Commerce untuk setiap kategori usia
data_ecommerce = {
    "Anak-Anak": 13,
    "Dewasa": 119,
    "Dewasa Muda": 44,
    "Lansia": 25,
    "Remaja": 22
}

# Menghitung rata-rata dan standar deviasi
mean_ecommerce = np.mean(list(data_ecommerce.values()))
std_dev_ecommerce = np.std(list(data_ecommerce.values()), ddof=1)  # ddof=1 untuk sample standard deviation
n_ecommerce = len(data_ecommerce)

# Menghitung Confidence Interval
z = norm.ppf(0.975)  # z-score untuk kepercayaan 95%
margin_of_error = z * (std_dev_ecommerce / np.sqrt(n_ecommerce))
lower_bound = mean_ecommerce - margin_of_error
upper_bound = mean_ecommerce + margin_of_error

# Menampilkan hasil
print("Rata-rata E-Commerce:", mean_ecommerce)
print("Standar Deviasi E-Commerce:", std_dev_ecommerce)
print("Confidence Interval (95%) untuk E-Commerce:", (lower_bound, upper_bound))
```
